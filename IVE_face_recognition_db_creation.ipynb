# IVE face recognition project

## FIRST STEP: Create the database that cotains the images for the IVE's members

For this step I choose SQLite because it doesn't requiere a previuos configuration and we can work through Python and a easy way.

Right now I have 33 images for each memeber and I will put that data into an SQL database

```python
import sqlite3

# Connection to SQLite database

db = sqlite3.connect(r'C:\Users\lonel\OneDrive\Escritorio\IVE face recognition\images.db')
cursor = db.cursor()

# Let's create the table to storage the images

cursor.execute("""
CREATE TABLE IF NOT EXISTS images (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    file_path TEXT NOT NULL,
    label TEXT NOT NULL
)
""")

# Commit the table creation

db.commit()

# Add the path for the imagen

ive_members = {
    "Gaeul": "G",
    "Leeseo": "Le",
    "Liz": "L",
    "Rei": "R",
    "Wonyoung": "W",
    "Yujin": "Y"
} # Here we map the members and the sintaxis for their image name format

# Apply a for loop that formats the url to add the photos to the database

for member, prefix in ive_members.items():
    for n in range(1, 34): # Because we have 33 images right now
        file_path = fr'C:\Users\lonel\OneDrive\Escritorio\IVE face recognition\{member}\{prefix}{n}.jpg'
        cursor.execute('INSERT INTO images (file_path, label) VALUES (?, ?)', (file_path, member))

db.commit()

# Check the database
cursor.execute("SELECT * FROM images")
print(cursor.fetchall())
db.close()
```
